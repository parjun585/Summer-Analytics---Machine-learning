### 2. How many cars have mpg  35?

```python
#your code here
num_cars = df[df['mpg'] >= 35].shape[0]
print(f"Number of cars with mpg >= 35: {num_cars}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
hp=df[(df['horsepower']>100) & (df['weight']<3000)]
hp
common_origin= hp['origin'].mode()[0]
print(f"The most common origin of cars with horsepower > 100 and weight < 3000 is: {common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
jpn_cars=df[df['origin'] == 'japan']
jpn_cars['acceleration'].mean().round(2)
print(f"The average mpg of Japanese cars is: {jpn_cars['acceleration'].mean().round(2)}")
```

### 5. Which year had the highest average mpg?

```python
#your code here
highest_acceleration = df.groupby('model_year')['mpg'].mean().idxmax()
print(f"The model year with the highest median mpg is: {highest_acceleration}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
# Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin
plt.figure(figsize=(12, 6))
df.groupby(['model_year', 'origin'])['mpg'].mean().unstack().plot(kind='line', marker='*')
plt.title('Average MPG Over the Years by Origin')
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import seaborn as sns

# Scatterplot of horsepower vs weight, colored by origin and sized by mpg
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="horsepower", y="weight", hue="origin", size="mpg",
                hue_order=['japan', 'europe', 'usa'], palette="Set1", edgecolor="black", alpha=1)

# Add titles and labels
plt.title("Relationship Between Horsepower and Weight", fontsize=14)
plt.xlabel("Horsepower", fontsize=12)
plt.ylabel("Weight", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```
